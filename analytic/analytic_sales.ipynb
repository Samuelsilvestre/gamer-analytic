```python
df_vendas = df.loc[(df['Global'] > 10)].reset_index()
df_vendas.head(10)
```

```python
# top 3 mais vendidos
percentual_vendas = df_vendas[['Genre','Global']].groupby('Genre').value_counts(normalize=True) * 100
percentual_vendas
```

```python
# percentual de vendas regiões
america_norte = [ round(america / total * 100) for america,total in zip(df_vendas['North America'], df_vendas['Global'])]
europa = [round(europa / total * 100) for europa,total in zip(df_vendas['Europe'], df_vendas['Global'])]
japao = [round(japao / total * 100) for japao,total in zip(df_vendas['Japan'], df_vendas['Global']) ]



```

```python
# criação de outro dataframe
dict_game = {
    'games': [game for game in df_vendas['Game']],
    'genere':[genre for genre in df_vendas['Genre']],
    'north america':america_norte,
    'europe':europa,
    'japao':japao,
    'global':[g for g in df_vendas['Global']]
}

df_game = pd.DataFrame(dict_game)
df_game.head(10)
```
